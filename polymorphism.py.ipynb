```python
# Create a class Animal with make_sound() and derived classes Dog, Cat, Cow that
#override it.
#Demonstrate polymorphism by iterating over a list of different animal objects and calling
#make_sound().
class Animal:
    def sound(self):
        print("m1")
class Dog(Animal):
    def sound(self):
        print("m2")
class Cat(Animal):
    def sound(self):
        print("m3")
class Cow(Animal):
    def sound(self):
        print("m4")
d=Dog()
c1=Cat()
c2=Cow()
animals=[d,c1,c2]
for animal in animals:
    animal.sound()
```

```python
#Write a function operate(device) that calls device.start().
#Pass in objects of Car, Computer, and WashingMachine — all of which define a start()
#method, but share no inheritance relationship.
#Show that Python’s polymorphism works through behavior, not type.

class Car:
    def start(self):
        print("m1")
class Computer:
    def start(self):
        print("m2")
class WashmingMachine:
    def start(self):
        print("m3")
def operate(device):
    device.start()
c=Car()
c1=Computer()
w=WashmingMachine()
operate(c)
operate(c1)
operate(w)
```

```python
#Create a Vector class that supports:
#• + operator → add coordinates
#• == operator → compare equality
#Show how operator overloading gives natural polymorphism to user-defined classes.

class Vector:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def __add__(self,other):
        return Vector(self.x+other.x,self.y+other.y)
    def __eq__(self,other):
        return self.x==other.x and self.y==other.y
    def __str__(self):
        return f"{self.x},{self.y}"
v1=Vector(2,3)
v2=Vector(3,7)
v3=Vector(8,23)
result=v1+v2
print(result)
print(v1==v2)
print(v2==v3)
```

```python
# Create a base class Transport with move() and derived classes Bus and Bike that
#override it but also call the parent implementation using super().
#Show the combination of reuse + custom behavior.
class Transport:
    def move(self):
        print("m1")
class Bus(Transport):
    def move(self):
        super().move()
        print("m2")
class Bike(Transport):
    def move(self):
        super().move()
        print("m3")
b1=Bus()
b2=Bike()
b1.move()
b2.move()
```

```python
#Using the abc module, create an abstract class Notification with send().
#Implement subclasses EmailNotification, SMSNotification, PushNotification — each
#with its own send() logic.
#Demonstrate polymorphism by looping over all and calling send()
from abc import ABC,abstractmethod
class Notifications(ABC):
    @abstractmethod
    def send(self):
        pass
class EmailNotifications(Notifications):
    def send(self):
        print("m1")
class SMSNotification(Notifications):
    def send(self):
        print("m2")
class PushNotifications(Notifications):
    def send(self):
        print("m3")
e = EmailNotifications()
s=SMSNotification()
p=PushNotifications()
e.send()
s.send()
p.send()
notifications=[e,p,s]
for i in notifications:
    i.send()
```
