```python
class loan:
    intrest = 0.1   # class variable

    def __init__(self, n, a):
        self.name = n
        self.amount = a

    def change_intrest(self, ni):
        loan.intrest = ni


l1 = loan("hi", 1000000)
l1.change_intrest(0.15)

print(loan.intrest)
```

```python
class payment:
    def __init__(self,amount):
        self.amount=amount
    @classmethod
    def upipayment(cls,a,p):
        print(f"upi payment to {p} is done")
        return cls(a)
p1=payment.upipayment(500,9999)
print(p1.amount)
```

```python
class Data:
    def __init__(self,d,m,y):
        self.d = d
        self.m = m
        self.y = y
    @classmethod
    def usdateformat(cls,date):
        m,d,y=date.split("/")
        return cls(d,m,y)
d1 =Data.usdateformat("01/06/2026")
print(d1.d)
```
