```python
#given a string, it should filter and contain only alphabets and refer sum of the ASCII value
def filter_and_sum_ascii(s):
    filtered = ''.join([i for i in s if i.isalpha()])
    ascii_sum = sum(ord(i) for i in filtered)
    return filtered, ascii_sum
text = "*@#HIBULLETt123"
filtered,total=filter_and_sum_ascii(text)
print(filtered)
print(total)
```

```python
#Given a list of numbers , remove duplicates , filter odd numbers,return highest odd number
def highest_odd_number(nums):
    unique_nums = set(nums)
    odd_nums = [i for i in unique_nums if i%2 != 0]
    return max(odd_nums) if odd_nums else None
numbers = [1,2,3,34,4,4,4,5,5,6,6,7,8,9,8,6,45,3]
result=highest_odd_number(numbers)
print(result)
```
