```python
#largest element in the array
def largest_numbers(nums):
    nums.sort()
    return nums[-1]
nums = [1,4,5,23,4,2,4]
print(largest_numbers(nums))
```

```python
#second largest number
def largest_number(nums):
    nums.sort()
    return nums[-2]
nums = [1,2,3,43,2,1234]
print(largest_number(nums))
```

```python
#selection sort:
def selection_sort(nums):
    n = len(nums)
    for i in range(n):
        min_val = i
        for j in range(i+1,n):
            if nums[j]<nums[min_val]:
                min_val=j
        nums[i],nums[min_val]=nums[min_val],nums[i]
    return nums
nums=[5,4,3,2,1]
print(selection_sort(nums))
```
